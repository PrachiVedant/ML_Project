```python
# UNIVARIATE ANALYSIS (GENDER DISTRIBUTION)
f,ax =plt.subplots(1,2,figsize = (20,10))
sns.countplot(x=df['gender'],data=df,palette='bright',saturation=0.95,ax=ax[0])
for container in ax[0].containers:
    ax[0].bar_label(container,color='black',size=20)

plt.pie(x=df['gender'].value_counts(),labels=['Male','Female'],explode=[0,0.1],autopct='%1.1f%%',shadow=True,colors=['#ff4d4d','#ff8000'])
plt.show()
```

```python
df['average_score'] = df[['math_score', 'reading_score', 'writing_score']].mean(axis=1)

# Group and calculate means
gender_group = df.groupby('gender')[['average_score', 'math_score']].mean()

# Fetch scores using .loc[]
female_scores = [
    gender_group.loc['female', 'average_score'],
    gender_group.loc['female', 'math_score']
]

male_scores = [
    gender_group.loc['male', 'average_score'],
    gender_group.loc['male', 'math_score']
]

# Plot
X = ['Total Average', 'Math Average']
X_axis = np.arange(len(X))

plt.figure(figsize=(10, 8))
plt.bar(X_axis - 0.2, male_scores, 0.4, label='Male', color='skyblue')
plt.bar(X_axis + 0.2, female_scores, 0.4, label='Female', color='lightcoral')
plt.xticks(X_axis, X)
plt.ylabel("Marks")
plt.title("Total Average vs Math Average by Gender", fontweight='bold')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
